### 1-4. 데이터 전처리

```python
# 기간은 날짜가 똑같기 때문에 무의미하다고 판단하여 삭제
df.drop('기간', axis = 1, inplace = True )
```

```python
df
```

- 합계와 소계가 들어있는 행 때문에 형변환이 계속 에러가 남 부득이하게 삭제
- 에러 = 만 단위, 천 단위에 (,)가 쓰여서 조건문, 삭제, 대체 등 다양하게 시도해도 오류가 해결이 안됨.

```python
# 합계와 소계가 포함된 행 삭제
df = df.loc[(~df['동'].str.contains('소계'))]
df=df.drop(0,axis=0)
```

```python
df
```

```python
# 자치구, 동, 합계는 int로 변환이 필요 없기 때문에 나머지 데이터들과 분리
df2 = df.drop( [ '자치구', '동', '합계' ], axis = 1 ) 
df2
```

```python
df2 = df2.apply(pd.to_numeric, errors = 'coerce').fillna(0)
```

```python
df2
```

```python
df2.dtypes
```

-> object타입을 성공적으로 int로 변환 함

```python
df3 = df.loc[:,['자치구','동','합계']]
df3
```

-> (자치구, 동, 합계) / 0세~100세 이상 데이터프레임을 다시 병합 해야함

```python
# 자치구, 동, 합계 데이터프레임과 연령별 데이터프레임 병합
df1 = pd.concat([df3, df2], axis=1)
df1
```

### 2. 데이터 기본 통계

```python
df2.describe()
```

### 2-1. 신생아 수 데이터 

- 문제 정의에 부합하는 데이터를 만들어서 표현

```python
# 자치구, 동만 분리하기
df3d = df3.drop(['합계'], axis=1)
df3d
```

```python
s1 = df2['0세']
s1
```

```python
df0 = pd.concat([df3d, s1], axis=1)
df0
```

- 자치구, 동별 0세 인구를 파악할 수 있는 데이터프레임을 구성 함.

```python
df0d = df0.describe()
df0d
```

- 자치구, 동별 0세 인구에 대한 데이터프레임의 기본통계를 확인

```python
# 기본 통계에서 평균 인구를 뽑음
df0d = df0d.iloc[1,0]
df0d
```

```python
print( '평균 신생아 수           : {:6.2f}명'.format( df0d ) )
print()
min_index = df0['0세'].idxmin()
max_index = df0['0세'].idxmax()
print( '최저 신생아 수 동네 : {:3d}명( 자치구: [{:>2s}]          [{:<5s})]'.format( df0.loc[ min_index, '0세' ], 
                                                                df0.loc[ min_index, '자치구' ], 
                                                                df0.loc[ min_index, '동' ] ) )
print()
print( '최고 신생아 수 동네 : {:3d}명( 자치구: [{:>2s}]          [{:<5s})]'.format( df0.loc[ max_index, '0세' ],
                                                                df0.loc[ max_index, '자치구' ],
                                                                df0.loc[ max_index, '동' ] ) )
print()
```

-> 신생아 수 데이터프레임을 바탕으로 나타낼 수 있는 자료들을 표현해 봄

#### 신생아 수 데이터프레임 시각화 

```python
dfb = df0.groupby(by=['자치구'], as_index=False).sum()
dfb
```

-> 시각화 표현을 위해서 400여개의 동을 변수로 하기엔 복잡하고 보기 힘든 시각화 자료가 만들어지기 때문에 각 자치구를
기준으로 그륩화 하여 자치구별 0세 인구에 대한 데이터프레임을 구성 함

```python
dfb.plot(kind = 'bar')
```

결과 = 가장 높은 신생아 인구수를 기록한 자치구는 송파구(17번)이라는 것이 확인 됨
반대로 가장 낮은 신생아 인구수를 기록한 자치구는 종로구(22번)임

### 2-2. 고령인구 수 데이터

- 고령인구를 80세~100세 이상으로 정의하여 고령인구를 파악하기로 함

```python
# 고령인구 80이상
df2s = df2[['80세','81세','82세','83세','84세','85세','86세','87세','88세','89세','90세','91세','92세','93세','94세','95세',
     '96세','97세','98세','99세','100세 이상+']]
df2s
```

- 80~100세 이상 인구에 대한 각 변수들을 분리하였고 각 열을 기준으로 데이터 값을 합하여 고령인구 수를 만들기로 함

```python
# 80세 이상 인구 합 구하기
df2s = df2s.sum(axis=1)
df2s
```

```python
dfs = pd.concat([df3d, df2s], axis= 1)
```

```python
dfs
```

- 고령인구수로 구성하기 위해 합쳐진 변수의 이름이 '0'으로 되어 있어서 알아보기 위해 '고령인구'로 이름을 붙여줌

```python
dfs.rename(columns = {0 : '고령인구'}, inplace = True)
```

```python
dfs
```

- 고령인구 수를 알아보기 위해 구성된 데이터프레임 완성

```python
dfs0 = dfs.describe()
dfs0
```

- 고령인구 수 데이터프레임에 대한 기본 통계, 평균~최댓값까지 잘 나타나 있음

```python
# 고령인구 수 기본통계에서 평균 값만 추출
dfs0 = dfs0.iloc[1,0]
dfs0
```

```python
print( '평균 고령인구            : {:6.2f}명'.format( dfs0 ) )
print()
min_index = dfs['고령인구'].idxmin()
max_index = dfs['고령인구'].idxmax()
print( '최저 고령인구 동네 : {:3d}명( 자치구: [{:>2s}]          [{:<5s})]'.format( dfs.loc[ min_index, '고령인구' ], 
                                                                dfs.loc[ min_index, '자치구' ], 
                                                                dfs.loc[ min_index, '동' ] ) )
print()
print( '최고 고령인구 동네 : {:3d}명( 자치구: [{:>2s}]          [{:<5s})]'.format( dfs.loc[ max_index, '고령인구' ],
                                                                dfs.loc[ max_index, '자치구' ],
                                                                dfs.loc[ max_index, '동' ] ) )
print()
```

- 고령인구 수 데이터프레임과 기본통계를 바탕으로 나타낼 수 있는 자료들을 표현해봄

```python
dfs1 = dfs.groupby(by=['자치구'], as_index=False).sum()
dfs1
```

-> 마찬가지로 고령인구 수 또한 각 자치구별 고령인구수를 그륩화하여 시각화를 하기 위해 고령인구 수를 그륩화하여 더함

```python
dfs1.plot(kind = 'bar')
```

-> 결과 = 가장 높은 고령인구 수를 기록한 자치구는 노원구(8번)이라는 것이 확인 됨
반대로 가장 낮은 고령인구 수를 기록한 자치구는 중구(23번)임
