```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame( { 'a': [ 'a0', 'a1', 'a2', 'a3' ],
                      'b': [ 'b0', 'b1', 'b2', 'b3' ],
                      'c': [ 'c0', 'c1', 'c2', 'c3' ] },
                      index = [ 0, 1, 2, 3] )
df1
```

```python
df2 = pd.DataFrame( { 'a': [ 'a2', 'a3', 'a4', 'a5' ],
                      'b': [ 'b2', 'b3', 'b4', 'b5' ],
                      'c': [ 'c2', 'c3', 'c4', 'c5' ],
                      'd': [ 'd2', 'd3', 'd4', 'd5' ] },
                      index = [ 2, 3, 4, 5 ] )
df2
```

```python
df = pd.concat( [df1, df2], axis = 1)
```

```python
df
```

```python
df0 = pd.concat( [df1, df2], ignore_index = True, axis = 1 )
```

```python
df0
```

```python
df = pd.concat([df1, df2], join = 'inner', axis = 1)
```

```python
df
```
